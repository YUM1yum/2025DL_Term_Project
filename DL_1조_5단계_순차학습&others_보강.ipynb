## 가상 데이터 생성

#### 1~4단계 가상 데이터 생성 & 3, 4단계 others label 병합
- 단계 올라갈수록 배경 데이터 복잡성 up
- 밝기에 따른 아이콘 배치
- 아이콘 배치 방식에 일렬 배치 추가
- 아이콘에 다양한 variation 주는 함수 추가
- 자연스러운 아이콘 첨부를 위한 함수 추가
- cvat으로 배경 데이터 라벨링한 라벨 병합

```python
import os
import random
import shutil
from PIL import Image, ImageFilter
import numpy as np
from pathlib import Path

# 설정
icon_total_min, icon_total_max = 10, 35
color_icon_max = 10
other_min, other_max = 7, 24
fixed_bg_size = (1280, 720)
brightness_threshold = 127
base_icon_size = 60

# 경로
icon_black_dir = Path("train_dataset/icons/black")
icon_white_dir = Path("train_dataset/icons/white")
icon_else_dir = Path("train_dataset/icons/else")
other_dir = Path("train_dataset/others")

black_icons = list(icon_black_dir.glob("**/*.png")) + list(icon_black_dir.glob("**/*.jpg"))
white_icons = list(icon_white_dir.glob("**/*.png")) + list(icon_white_dir.glob("**/*.jpg"))
color_icons = list(icon_else_dir.glob("**/*.png")) + list(icon_else_dir.glob("**/*.jpg"))
other_images = list(other_dir.glob("**/*.png")) + list(other_dir.glob("**/*.jpg"))

assert black_icons and white_icons, "❌ 흑백 아이콘 없음"
assert other_images, "❌ others 없음"

# 유틸 함수
def is_bright(patch):
    gray = np.array(patch.convert("L"))
    return np.mean(gray) > brightness_threshold

def resize_keep_ratio(img, target_long):
    w, h = img.size
    scale = target_long / max(w, h)
    return img.resize((int(w * scale), int(h * scale)), Image.LANCZOS)

def is_overlap(box1, box2):
    x1_min, y1_min, x1_max, y1_max = box1
    x2_min, y2_min, x2_max, y2_max = box2
    return not (x1_max < x2_min or x2_max < x1_min or y1_max < y2_min or y2_max < y1_min)

def apply_effects(img):
    if random.random() < 0.5:
        angle = random.randint(-30, 30)
        img = img.rotate(angle, expand=True)
    if random.random() < 0.3:
        alpha = img.split()[-1].point(lambda p: p * random.uniform(0.3, 0.8))
        img.putalpha(alpha)
    if random.random() < 0.3:
        img = img.filter(ImageFilter.GaussianBlur(radius=random.uniform(0.5, 1.5)))
    return img

def blend_icon(bg, icon, x, y):
    temp = Image.new("RGBA", bg.size)
    temp.paste(icon, (x, y), icon)
    return Image.alpha_composite(bg.convert("RGBA"), temp)

def place_icons_mixed(src_list, class_id, scale_range, total_count, img_w, img_h, bg, placed_boxes, label_lines, line_ratio=0.7):
    count_line = int(total_count * line_ratio)
    count_random = total_count - count_line

    # 일렬 배치
    x = random.randint(10, 100)
    y_line = random.randint(30, int(img_h * 0.3))
    for _ in range(count_line):
        for _ in range(20):
            src = random.choice(src_list)
            icon = Image.open(src).convert("RGBA")
            icon = apply_effects(icon)
            scale = random.uniform(*scale_range)
            icon = resize_keep_ratio(icon, int(base_icon_size * scale))
            w, h = icon.size
            if x + w > img_w:
                break
            y = y_line + random.randint(-5, 5)
            box = (x, y, x + w, y + h)
            if any(is_overlap(box, b) for b in placed_boxes):
                x += w + random.randint(5, 25)
                continue
            bg = blend_icon(bg, icon, x, y)
            placed_boxes.append(box)
            label_lines.append(f"{class_id} {(x+w/2)/img_w:.6f} {(y+h/2)/img_h:.6f} {w/img_w:.6f} {h/img_h:.6f}")
            x += w + random.randint(5, 25)
            break

    # 랜덤 배치
    for _ in range(count_random):
        for _ in range(20):
            src = random.choice(src_list)
            icon = Image.open(src).convert("RGBA")
            icon = apply_effects(icon)
            scale = random.uniform(*scale_range)
            icon = resize_keep_ratio(icon, int(base_icon_size * scale))
            w, h = icon.size
            x = random.randint(0, img_w - w)
            y = random.randint(0, img_h - h)
            box = (x, y, x + w, y + h)
            if any(is_overlap(box, b) for b in placed_boxes):
                continue
            bg = blend_icon(bg, icon, x, y)
            placed_boxes.append(box)
            label_lines.append(f"{class_id} {(x+w/2)/img_w:.6f} {(y+h/2)/img_h:.6f} {w/img_w:.6f} {h/img_h:.6f}")
            break

    return bg, placed_boxes, label_lines

# ✅ Stage 1~4 synthetic 생성
for stage in range(1, 5):
    print(f"\n🔄 Stage {stage} 생성 시작")
    bg_dir = Path(f"train_dataset/stage{stage}")
    out_img_dir = Path(f"train_dataset/stage{stage}_enhanced/images/train")
    out_lbl_dir = Path(f"train_dataset/stage{stage}_enhanced/labels/train")
    out_img_dir.mkdir(parents=True, exist_ok=True)
    out_lbl_dir.mkdir(parents=True, exist_ok=True)

    bg_images = list(bg_dir.glob("*.png")) + list(bg_dir.glob("*.jpg"))
    assert bg_images, f"❌ Stage {stage} 배경 없음"

    for idx in range(400):
        bg = Image.open(random.choice(bg_images)).convert("RGB").resize(fixed_bg_size)
        img_w, img_h = bg.size
        label_lines = []
        placed_boxes = []

        # others
        for _ in range(random.randint(other_min, other_max)):
            bg, placed_boxes, label_lines = place_icons_mixed(
                other_images, 1, (0.5, 1.2), 1,
                img_w, img_h, bg, placed_boxes, label_lines,
                line_ratio=0.0
            )

        # color + gray 아이콘 전체 합쳐서 배치
        total_icons = random.randint(icon_total_min, icon_total_max)
        color_count = random.randint(0, min(color_icon_max, total_icons))
        gray_count = total_icons - color_count

        if color_count > 0:
            bg, placed_boxes, label_lines = place_icons_mixed(
                color_icons, 0, (0.45, 0.58), color_count,
                img_w, img_h, bg, placed_boxes, label_lines
            )

        for _ in range(gray_count):
            patch_w = patch_h = base_icon_size
            x = random.randint(0, img_w - patch_w)
            y = random.randint(0, img_h - patch_h)
            patch = bg.crop((x, y, x + patch_w, y + patch_h))
            bright = is_bright(patch)
            candidates = black_icons if bright else white_icons
            bg, placed_boxes, label_lines = place_icons_mixed(
                candidates, 0, (0.3, 1.0), 1,
                img_w, img_h, bg, placed_boxes, label_lines,
                line_ratio=0.0
            )

        img_path = out_img_dir / f"{idx:04d}.jpg"
        lbl_path = out_lbl_dir / f"{idx:04d}.txt"
        bg.convert("RGB").save(img_path)
        with open(lbl_path, "w") as f:
            f.write("\n".join(label_lines))

    print(f"✅ Stage {stage} synthetic 생성 완료")

# ✅ Stage 3, 4: others 라벨링 병합
def merge_manual_others(stage):
    print(f"\n📦 Stage {stage} others 라벨링 데이터 병합 중...")
    manual_img_dir = Path(f"train_dataset/stage{stage}")
    manual_lbl_dir = Path(f"train_dataset/stage{stage}_label")
    target_img_dir = Path(f"train_dataset/stage{stage}_enhanced/images/train")
    target_lbl_dir = Path(f"train_dataset/stage{stage}_enhanced/labels/train")

    count = 0
    for img_path in manual_img_dir.glob("*"):
        if img_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
            continue
        lbl_path = manual_lbl_dir / (img_path.stem + ".txt")
        if lbl_path.exists():
            shutil.copy(img_path, target_img_dir / img_path.name)
            shutil.copy(lbl_path, target_lbl_dir / lbl_path.name)
            count += 1
        else:
            print(f"⚠️ 라벨 없음 → {img_path.name}")

    print(f"✅ Stage {stage} 병합 완료! 수동 others 라벨 {count}장 추가됨")

# 병합 실행
merge_manual_others(3)
merge_manual_others(4)
```

#### 5단계 가상 데이터 생성
- 다양한 edge case
- epoch, 생성 이미지 수, patience 조정으로 과적합 방지

```python
import os
import shutil
import random
from PIL import Image, ImageFilter
import numpy as np
from pathlib import Path

# 경로 및 설정
base_icon_size = 60
bg_source = Path("train_dataset/stage4")
bg_target = Path("train_dataset/stage5")
img_out = Path("train_dataset/stage5_enhanced/images/train")
lbl_out = Path("train_dataset/stage5_enhanced/labels/train")
img_out.mkdir(parents=True, exist_ok=True)
lbl_out.mkdir(parents=True, exist_ok=True)
bg_target.mkdir(parents=True, exist_ok=True)

# 아이콘 경로
icon_black_dir = Path("train_dataset/icons/black")
icon_white_dir = Path("train_dataset/icons/white")
icon_color_dir = Path("train_dataset/icons/else")
other_dir = Path("train_dataset/others")

# 복사: stage4 → stage5_bg
for img_file in list(bg_source.glob("*.jpg")) + list(bg_source.glob("*.png")):
    shutil.copy(img_file, bg_target / img_file.name)

# 이미지 리스트
bgs = list(bg_target.glob("*.jpg")) + list(bg_target.glob("*.png"))
black_icons = list(icon_black_dir.glob("*"))
white_icons = list(icon_white_dir.glob("*"))
color_icons = list(icon_color_dir.glob("*"))
other_icons = list(other_dir.glob("*"))

# 유틸
def resize_keep_ratio(img, target_long):
    w, h = img.size
    scale = target_long / max(w, h)
    return img.resize((int(w * scale), int(h * scale)), Image.LANCZOS)

def blend_icon(bg, icon, x, y):
    temp = Image.new("RGBA", bg.size)
    temp.paste(icon, (x, y), icon)
    return Image.alpha_composite(bg.convert("RGBA"), temp)

def apply_effects(icon):
    if random.random() < 0.3:
        icon = icon.rotate(random.randint(-30, 30), expand=True)
    if random.random() < 0.3:
        icon = icon.filter(ImageFilter.GaussianBlur(radius=random.uniform(0.5, 1.5)))
    return icon

def place_icon(bg, icon, class_id, size_scale, label_lines):
    icon = apply_effects(icon)
    icon = resize_keep_ratio(icon, int(base_icon_size * size_scale))
    w, h = icon.size
    img_w, img_h = bg.size
    x = random.randint(0, img_w - w)
    y = random.randint(0, img_h - h)
    bg = blend_icon(bg, icon, x, y)
    label_lines.append(f"{class_id} {(x+w/2)/img_w:.6f} {(y+h/2)/img_h:.6f} {w/img_w:.6f} {h/img_h:.6f}")
    return bg

# ✅ 특이 케이스 포함 100장 생성
for idx in range(100):
    bg = Image.open(random.choice(bgs)).convert("RGB").resize((1280, 720))
    label_lines = []

    mode = idx % 6

    if mode == 0:
        # ✅ 아이콘 없음
        pass

    elif mode == 1:
        # ✅ 아이콘만 있음
        for _ in range(random.randint(10, 20)):
            icon = Image.open(random.choice(color_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=0, size_scale=random.uniform(0.4, 1.0), label_lines=label_lines)

    elif mode == 2:
        # ✅ others만 있음
        for _ in range(random.randint(5, 15)):
            icon = Image.open(random.choice(other_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=1, size_scale=random.uniform(0.5, 1.2), label_lines=label_lines)

    elif mode == 3:
        # ✅ 아이콘 크게
        for _ in range(3):
            icon = Image.open(random.choice(color_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=0, size_scale=random.uniform(2.0, 3.0), label_lines=label_lines)

    elif mode == 4:
        # ✅ 아이콘 작게
        for _ in range(20):
            icon = Image.open(random.choice(color_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=0, size_scale=random.uniform(0.1, 0.2), label_lines=label_lines)

    else:
        # ✅ 아이콘 + others 혼합
        for _ in range(random.randint(5, 15)):
            icon = Image.open(random.choice(color_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=0, size_scale=random.uniform(0.4, 1.0), label_lines=label_lines)
        for _ in range(random.randint(3, 10)):
            icon = Image.open(random.choice(other_icons)).convert("RGBA")
            bg = place_icon(bg, icon, class_id=1, size_scale=random.uniform(0.5, 1.2), label_lines=label_lines)

    # 저장
    bg.convert("RGB").save(img_out / f"{idx:04d}.jpg")  # ← 여기 `.convert("RGB")` 추가!
    with open(lbl_out / f"{idx:04d}.txt", "w") as f:
        f.write("\n".join(label_lines))

print("✅ Stage5 특이 케이스 100장 생성 완료")
```

```python
from pathlib import Path

for stage in range(1, 6):
    yaml_path = Path(f"train_dataset/stage{stage}_enhanced/data.yaml")
    yaml_path.parent.mkdir(parents=True, exist_ok=True)
    with open(yaml_path, "w") as f:
        f.write(
            f"train: train_dataset/stage{stage}_enhanced/images/train\n"
            f"val: train_dataset/stage{stage}_enhanced/images/train\n\n"
            f"nc: 2\n"
            f"names: [\"icon\", \"other\"]\n"
        )
print("✅ data.yaml 생성 완료")
```

```python
from ultralytics import YOLO
from pathlib import Path

base_model = 'yolov8s.pt'
epochs_list = [20, 40, 40, 60, 10]
root = '/content'  # 🔥 colab root

for stage in range(1, 6):
    data_yaml = f'{root}/train_dataset/stage{stage}_enhanced/data.yaml'
    project_dir = f'{root}/icon_detector'
    run_name = f'stage{stage}'

    print(f'\n🚀 Stage {stage} 학습 시작...')

    if stage == 1:
        model = YOLO(base_model)
    else:
        prev_best = Path(project_dir) / f'stage{stage-1}' / 'weights' / 'best.pt'
        model = YOLO(str(prev_best))

    model.train(
        data=data_yaml,
        epochs=epochs_list[stage - 1],
        imgsz=1280,
        rect=True,
        project=project_dir,
        name=run_name,
        exist_ok=True,
        patience=5,
        cos_lr=True,
        lr0=0.01,
        warmup_epochs=5,
        weight_decay=0.001,
        iou=0.7,
        conf=0.4,
        augment=True
    )
```
